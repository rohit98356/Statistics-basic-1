```python
Q2. What are the measures of central tendency, and when should you see each?
discuss the mean, median, and mode with examples and situations where each is appropriate. 
```

```python
Measures of Central Tendency

. measures of central tendency describe a dataset's center or typical value. 
the three main measures are mean, median, and mode.
each provides unique isights and is appropriate in different scenarios depending
on the data's nature and distribution.
```

```python
import numpy as np
from scipy import stats

data = [12, 15, 12, 10, 14, 13, 15, 12, 17, 10]

mean = np.mean(data)

median = np.median(data)

mode_result = stats.mode(data)

if isinstance(mode_result.mode, np.ndarray):
    mode_value = mode_result.mode[0]
    mode_frequency = mode_result.count[0]

else:
    mode_value = mode_result.mode
    mode_frequency = mode_result.count

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode_value, "with frequency:", mode_frequency)
```

```python
Concept of Dispersion

. Dispersion refers to the extend to which data points in a dataset very or spread around a central value.
it provides insight  into the variability or  consistency of the data.
high dispersion indicates data points are spread out, 
while low dispersion implies they are clustered close together.

Variance

    . variance measures the average sequared difference between each data point and the mean.
it gives a sense of how far the data points deviate from the mean on average.
```

```python
import numpy as np 
import pandas as pd 

data = [4, 6, 8, 10, 12]

mean = np.mean(data)

population_variance = np.var(data)
sample_variance = np.var(data, ddof=1)

population_std_dev = np.std(data)
sample_std_dev = np.std(data, ddof=1)

print(f"Data: {data}")
print(f"Mean: {mean}")
print(f"Population Variance: {population_variance}")
print(f"Sample Variance: {sample_variance}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Sample Standard Deviation: {sample_std_dev}")
```

```python
Skewness-
          skewness is a statisticall measure that describe the degree of asymmetry of a distribution 
around its mean.

in a perfectly symmetrical distribution , the left and right sides are mirror images.
However, when the distribution is not symmetrical, it is said to be skewed.

    TYPES OF SKEWNESS-
1- Positive skewness(Right- skewed Distribution)
2- Negative Skewness(Left-skewed Distribution)
3- Zero Skewness(Symmetrical Distribution)

    HOW SKEWNESS AFFECTS THE INTERPRETATION OF DATA

1- Central Tendency Measures:
    . in skewed distributions , the mean is pulled in the direction of the tail,
making it less representative of the central tendency.
the median is a mor robust measure in such cases.

2- Outlier Influence:

. Skewness indicates the presence and influence of outliers.
positive skewness suggests outliers on the higher end, while negative skewness indicates outliers on the lower end.

3- Predictive Modeling:

. Skewed data may affect the performance of models that assume normality.
Transformations are often used to address skewness.

4- Decision - making:

. Understanding skewness helps in making informed decisions, especially in financial,
quality control, and risk management scenarios, where exterme values can have a significant impact.

5- Data Transformation Needs:

. Highly skewed data often requires normalization or standardization for better 
interpretability or compatibility with statistical techniques.
```

```python
the interquartile range(IQR) 
is a measure  if statistical dispersion, or how spread out the data points are in a dataset.
it is the range between the first quatile(Q1)
and the third quartile(Q3) of a dataset.
in simple terms, the IQR tells you how the middle 50% of your data is spread out.

1- Q1- 
the median of the lower half of the data (25th percentile).


2- Q3-
                                          the median of the upper half of the data (75th percentile).
```

```python
import numpy as np
import pandas as pd
data = [1, 2, 3, 4, 100, 200, 3, 4,5 ]
data_series = pd.Series(data)
Q1 = np.percentile(data_series, 25)
Q3 = np.percentile(data_series, 75)

IQR = Q3 -Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data_series[(data_series < lower_bound) | (data_series > upper_bound)]
print(f"Data: {data}")
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {list(outliers)}")
```

```python
Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7rule).
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0
sigma = 1

x = np.linspace(mu - 4*sigma,  mu + 4*sigma,  1000)

y = norm.pdf(x, mu, sigma)

plt.fill_between(x, y, where=(x >= mu - sigma) & (x <= mu + sigma), color='yellow', alpha=0.5, label='68% area')
plt.fill_between(x, y, where=(x >= mu - 2*sigma) & (x <= mu + 2*sigma), color='green', alpha=0.5, label='95% area')
plt.fill_between(x, y, where=(x >= mu - 3*sigma) & (x <= mu + 3*sigma), color='red', alpha=0.5, label='99.7% area')

plt.title("Normal Distribution and Empirical Rule", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
Q10. provide a real-life example of a poisson process and calculate the  probability for a specific event.
```

```python
import math
lambda_rate = 5
k = 3

poisson_probability = (lambda_rate**k * math.exp(-lambda_rate)) / math.factorial(k)

print(f"The probability of receiving exactly {k} calls in a hour is: {poisson_probability:.4f}")

cumulative_probability = sum((lambda_rate**i * math.exp(-lambda_rate)) / math.factorial(i) for i in range(k + 1))

print(f"The cumulative probability of receiving up to {k} calls in an hour is: {cumulative_probability:.4f}")

```

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
Y = np.array([50, 55, 60, 65, 70])

mean_X = np.mean(X)
mean_Y = np.mean(Y)

deviation_X = X - mean_X
deviation_Y = Y - mean_Y

product_deviations = deviation_X * deviation_Y

sum_product_deviations = np.sum(product_deviations)

covariance = sum_product_deviations / len(X)

covariance
```
